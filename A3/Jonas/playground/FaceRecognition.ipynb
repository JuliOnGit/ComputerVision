```
def useHumanForDecision(img1, img2):
  plt.imshow(img1)
  plt.show()
  plt.imshow(img2)
  plt.show()
  result = input("Same person (enter 1) or different (enter 0): ") 
  return result
```

```
f = open('testPairs', 'r')
output = open('results.txt', 'w')
lines = 0
for x in f:
  pair = x.split()
  lines += 1
  first_image = face_recognition.load_image_file("test/{}".format(pair[0]))
  second_image = face_recognition.load_image_file("test/{}".format(pair[1]))

  first_encodings = face_recognition.face_encodings(first_image)
  if first_encodings:
    first_encoding = first_encodings[0]
  else:
    print('Something failed on picture: {}'.format(pair[0]))
    result = useHumanForDecision(first_image, second_image)
    output.write('{}.0\n'.format(result))
    continue

  second_encodings = face_recognition.face_encodings(second_image)
  if second_encodings:
    second_encoding = second_encodings[0]
  else:
    print('Something failed on picture: {}'.format(pair[1]))
    result = useHumanForDecision(first_image, second_image)
    output.write('{}.0\n'.format(result))
    continue


  results = face_recognition.compare_faces([first_encoding], second_encoding)
  if results == [True]:
    print('Pictures are from the same person')
    output.write('1.0\n')
  else:
    print('Pictures are not from the same person')
    output.write('0.0\n')
f.close()
output.close()
print('Succesfully compared {} pictures'.format(lines))
```
